### Plotting and rendering features
Next example shows the features of drawing and plotting that can be used in Jupyter notebooks. Before that, we first need to `import` the required libraries that render the plots.

```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib
```

```python
# Data for plotting

# Create an array 't' for the x-axis. 
# It represents time, going from 0.0 to 2.0 seconds in steps of 0.01 seconds.
t = np.arange(0.0, 2.0, 0.01)

# Create an array 's' for the y-axis.
# This calculates a sine wave with a DC offset: s = 1 + sin(2 * pi * t).
# This will represent the voltage signal.
s = 1 + np.sin(2 * np.pi * t)

# Create a figure and a set of subplots. 
# 'fig' is the overall window or page, and 'ax' is the plot itself.
fig, ax = plt.subplots()

# Plot the data on the axes 'ax'.
# 't' is used for the x-coordinates and 's' is used for the y-coordinates.
ax.plot(t, s)

# Set the labels for the axes and the title for the plot.
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Plotting example')

# Add a grid to the background of the plot for better readability.
ax.grid()

# To display the plot, you would typically add:
plt.show()
```
